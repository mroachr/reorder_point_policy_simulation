# Introduction

This notebook shows the effects of using a reorder point policy and how a basic discrete event simulation model can be used to obtain the probability of a stockout situation. Although the model itself is rather simple, it would be easy to incorperate more complex functionallities, e.g., holding cost, punish cost, order cost, etc.

For this example, consider a system that represents the inventory level for one spesific product. The system use a basic reorder point policy (ROP), meaning that a new order is issued once the inventory level for said product reaches a certian level, often denoted as `r`. Furthermore, for each order a fixed quantity is ordered, often denoted as `Q`. In this example, r = 7 and Q = 10. The consumption of inventory is following a poisson process (lambda = 1 per time unit) and the lead time for new supplies is following an Erlang process (k = 4 and lambda = 2). Since the mean of an Erlang process is found by k / lambda, this means that the mean lead time for supplies are 2 time units, and when an order is placed there is still 7 units to go before we run into a stockout. The question is then: will we sill run into a stockout? By just looking at the mean lead time for suppies, this does not seem to be the case, but let's find out.

# Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import erlang
import random
import math
```

# Describing the poisson process

```python
def poisson_PMF(lam, k):
  return (lam ** k) * math.exp(-lam) / math.factorial(k)

k = [i for i in range(0, 15)]
P = [poisson_PMF(lam=1, k=k[i]) for i in range(len(k))]
plt.figure(figsize=(10, 7))
plt.plot(k, P, marker='o', markerfacecolor='darkorange', 
         markersize=10, color='dimgray')
plt.title('Poisson PMF')
plt.xlabel('k')
plt.ylabel('P(x=k)')
plt.show()
```

When using a poisson *process*, the `np.random.poisson(1)` function can also be used.

# Describing the Erlang process

```python
def erlang_PDF(lam, k, x):
  return (lam ** k) * (x ** (k - 1)) * math.exp(-lam * x) / math.factorial(k - 1)

k = 4
lam = 2
x = [i for i in range(0, 15)]
P = [erlang_PDF(lam, k, x[i]) for i in range(len(x))]

plt.figure(figsize=(10, 7))
plt.plot(x, P, marker='o', markerfacecolor='darkorange', 
         markersize=10, color='dimgray')
plt.title('Erlang (4) PDF')
plt.xlabel('x')
plt.ylabel('P')
plt.show()
```

When using a Erlang (k=4) *process*, the `random.gammavariate(4, 1)` function can also be used.

# Creating a simulation model

```python
class Simulation():
  def __init__(self, r, Q, inventory_max, inventory_min):
    # Variables/settings
    self.r = r
    self.Q = Q
    self.inventory_max = inventory_max
    self.inventory_min = inventory_min
    
    # Initialize dependant variables
    self.inventory = self.inventory_max
    self.t_consume = self.generate_consume()
    self.t_supply = float('inf')
    self.time = 0
    self.stockout_time = 0
    self.n_orders = 0
    self.n_orders_with_stockout = 0

  
  def advance_time(self):
    # Note current system state, for later use
    time_before_advancement = self.time
    inventory_before_advancement = self.inventory

    # Update time variable
    self.time = min(self.t_consume, self.t_supply)

    # Handle the next event
    if self.t_consume < self.t_supply:
      self.handle_consume_event()
    else:
      self.handle_supply_event()
      if inventory_before_advancement <= 0:
        self.n_orders_with_stockout += 1

    # If previous inventory was below 0, update time with stockout
    if inventory_before_advancement <= 0:
      self.stockout_time += self.time - time_before_advancement

  def handle_consume_event(self):
    # Update inventory
    self.inventory -= 1

    # Get next consume event
    if self.inventory > self.inventory_min:
      self.t_consume = self.time + self.generate_consume()
    else:
      self.t_consume = float('inf')

    # Generate new consume event, if needed and one does not already exist
    if self.inventory <= self.r and self.t_supply == float('inf'):
      self.t_supply = self.time + self.generate_supply()

  
  def handle_supply_event(self):
    # If t_consume is locked (due to inv_min has been reached), unlock t_consume
    if self.inventory == self.inventory_min:
      self.t_consume = self.time + self.generate_consume()

    # Update inventory
    self.inventory += self.Q

    # If inventory is still below r, 
    if self.inventory <= self.r:
      self.t_supply = self.time + self.generate_supply()
    else:
      self.t_supply = float('inf')

  
  def generate_consume(self):
    return np.random.poisson(1)

  
  def generate_supply(self):
    self.n_orders += 1
    return int(random.gammavariate(4, 1))
```

# Running the simulation model

```python
# Set seed
random.seed(42)
np.random.seed(42)

# Initialize dataframe
df = pd.DataFrame(columns=['time', 'inventory', 't_consume', 't_supply', 'stockout_time', 'n_orders', 'n_orders_with_stockout'])

# Parameters
r = 7                  # Re-order point (ROP)
Q = 10                 # Order quantity 
inventory_max = 17     # Max inventory capacity
inventory_min = -20    # Max backlog allowed
n = 10000              # Number of events to run

# Create simulation
s = Simulation(r, Q, inventory_max, inventory_min)

# Advance time n times
for step in range(n):
  s.advance_time()
  df.loc[len(df.index)] = [s.time, s.inventory, s.t_consume, s.t_supply, s.stockout_time, s.n_orders, s.n_orders_with_stockout]
```

# Results

```python
# Calculate probability of stockout
P_stockout_per_day = round(s.stockout_time / s.time, 5)
P_stockout_per_order = round(s.n_orders_with_stockout / s.n_orders, 5)

P_stockout_per_day, P_stockout_per_order
```

For an arbitrary time unit, the probability of having a stockout situation is 2.3%, and the probability of an arbitrary order will not arrive in time is 10.56%. Certainly a lot higher than one may imagine just by processing the information intuitively.

```python
start = 0
stop = 1000

plt.figure(figsize=(15, 7))

plt.plot(df[start:stop].time, df[start:stop].inventory, drawstyle='steps-pre', label='Inventory')
plt.hlines(y=r, xmin=0, xmax=df[start:stop]['time'].iloc[-1], color='red', label='Re-order point (ROP)')
plt.hlines(y=0, xmin=0, xmax=df[start:stop]['time'].iloc[-1], color='darkgrey', label='Stock-out')

plt.legend()
plt.title('Inventory level vs. time')
plt.ylabel('Inventory')
plt.xlabel('Time units')
plt.ylim(inventory_min - 1, inventory_max + 1)
plt.show()
```

The figure, although only showing a small part of the full timeline, illustrates how the inventory level changes over time.
